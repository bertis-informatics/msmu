# 04_summarize_multiset


```python
import msmu as mm

mdata = mm.read_h5mu("./data/h5mu/04_re_mapped.h5mu")
mdata
```

```python
mdata = mm.pp.log2_transform(mdata=mdata, level="psm")
mdata = mm.pp.normalise(mdata=mdata, method="quantile", level="psm", fraction=True)
mdata = mm.pp.scale_data(mdata=mdata, method="gis", level="psm", gis_prefix="POOLED")
```

```python
mdata = mm.pp.to_peptide(mdata=mdata, protein_col="proteins")
mdata = mm.pp.to_protein(mdata=mdata, protein_col="proteins")
mdata
```

```python
# Write h5mu file
mdata.write("./data/h5mu/05_summarized.h5mu")
```
