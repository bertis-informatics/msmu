```python
import msmu as mm
import pandas as pd
```

```python
annot_df = pd.read_csv("./data/experiment_annotation.tsv", sep="\t", dtype=str)
annot_df
```

```python
sage_dirs = {
    "S1": "./data/sage/S1",
    "S2": "./data/sage/S2",
}

mdatum = {}
for key, sage_dir in sage_dirs.items():
    mdata = mm.rw.read_sage(
        sage_output_dir=sage_dir,
        sample_name=annot_df.loc[annot_df["set"] == key, "sample_name"].tolist(),
        channel=annot_df.loc[annot_df["set"] == key, "channel"].tolist(),
        label="tmt",
    )

    mdatum[key] = mdata
```
