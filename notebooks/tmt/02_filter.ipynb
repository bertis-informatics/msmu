# 02_filter


```python
import msmu as mm
```

## Marking


```python
# mzml_dir = "PATH_TO_MZML_FILES"
mdata = mm.read_h5mu("./data/h5mu/01_read_search_output.h5mu")
mdata
```

```python
for mod in mdata.mod_names:
    mdata = mm.pp.add_all_nan_filter(mdata=mdata, modality=mod)
    mdata = mm.pp.add_prefix_filter(mdata=mdata, modality=mod, prefix=("rev_", "contam_"))
    mdata = mm.pp.add_q_value_filter(mdata=mdata, modality=mod, threshold=0.01)
    # mdata = mm.pp.add_precursor_purity_filter(
    #     mdata=mdata,
    #     modality=mod,
    #     threshold=0.7,
    #     mzml_files=[f for f in Path(mzml_dir).glob("*.mzML")],
    #     n_cores=6,
    # )

mdata
```

```python
# Write h5mu file
mdata.write("./data/h5mu/02_marked.h5mu")
```

## Filtering


```python
mdata = mm.pp.apply_filter(mdata=mdata, modality=mdata.mod_names)
mdata = mdata[mdata.obs_names.str.contains("BLANK") == False].copy()
mdata
```

```python
# Write h5mu file
mdata.write("./data/h5mu/03_filtered.h5mu")
```
